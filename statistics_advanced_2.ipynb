## **Q1. What is Hypothesis testing in statistics?**
Hypothesis testing is a statistical method used to make inferences about a population parameter based on a sample. It involves formulating two opposing hypotheses — the null hypothesis (H₀) and the alternative hypothesis (H₁) — and using sample data to decide which is more plausible. The process determines whether observed results are due to random chance or represent a statistically significant effect.

## **Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

  - Null Hypothesis (H₀): Assumes no effect or no difference exists (status quo).

  - Alternative Hypothesis (H₁): States that an effect or difference exists.

For example, if testing whether a new drug lowers blood pressure:

  - H₀: The drug has no effect (mean difference = 0).

  - H₁: The drug reduces blood pressure (mean difference < 0).

The difference is that H₀ is the default assumption, while H₁ challenges it.

## **Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

The significance level (α) is the threshold probability of rejecting the null hypothesis when it is true. Commonly set at 0.05 (5%), it represents the acceptable risk of a Type I error.

 - If the p-value ≤ α, reject H₀ (evidence supports H₁).

 - If the p-value > α, fail to reject H₀ (not enough evidence against it).

Thus, α controls how strong the evidence must be to reject H₀.

## **Q4. What are Type I and Type II errors? Give examples of each.**
 - Type I Error (False Positive): Rejecting H₀ when it is true.

      Example: Concluding a drug works when it actually does not.

 - Type II Error (False Negative): Failing to reject H₀ when H₁ is true.

      Example: Concluding a drug doesn’t work when it actually does.

## **Q5. What is the difference between a Z-test and a T-test? Explain when to use each.**
 - Z-test: Used when the population standard deviation (σ) is known, and the sample size is large (n > 30).

 - T-test: Used when σ is unknown, and we estimate it using the sample standard deviation (s). Suitable for small samples (n ≤ 30).

Example:

 - Checking if a sample mean differs from a known population mean with large n → Z-test.

 - Comparing two sample means with small n → T-test.

## **Q6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n, p = 10, 0.5
size = 1000

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor="black", align="left")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

## **Q7.Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.**
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu0 = 50
mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z_stat = (mean - mu0) / (std/np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

## **Q8.Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate data
data = np.random.normal(loc=100, scale=15, size=1000)

# Calculate mean & CI
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.norm.interval(0.95, loc=mean, scale=sem)

print("Mean:", mean)
print("95% Confidence Interval:", ci)

# Plot
plt.hist(data, bins=30, edgecolor="black", alpha=0.7)
plt.title("Normal Distribution Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(ci[0], color='red', linestyle='dashed', linewidth=2)
plt.axvline(ci[1], color='red', linestyle='dashed', linewidth=2)
plt.show()
```

## **Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample dataset
data = np.random.normal(50, 10, 100)

# Calculate Z-scores
z_scores = stats.zscore(data)

# Plot histogram
plt.hist(z_scores, bins=20, edgecolor="black")
plt.title("Z-scores Histogram")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

print("Z-scores:", z_scores[:10])  # preview first 10
```
